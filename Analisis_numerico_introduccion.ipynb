##Fibonacci:
Función que calcule los valores de la sucesión de Fibonacci y a que orden de complejidad pertenece O(?), JUSTIFICA TU RESPUESTA.
ejemplo: fibonacci(6) devuelve 8

```python
"""El código que proporcionó implementa la secuencia de Fibonacci de forma recursiva."""
"""La fibonaccifunción toma un número entero ncomo entrada.
  Si nes menor o igual a 1, la función devuelve n.
  Si nes mayor que 1, la función se llama a sí misma dos veces con n-1y n-2como argumentos y
  devuelve la suma de los dos resultados.
  La printdeclaración al final llama a la fibonaccifunción con un argumento de 6 e imprime el resultado.
  Entonces, cuando ejecuta el código, debe generar el sexto número en la secuencia de Fibonacci, que es 8."""

def fibonacci(n):


    if n<= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(6))
```

##Inversa de una matriz:
La inversa de una matriz de 2x2 y a que orden de complejidad pertenece O(?), JUSTIFICA TU RESPUESTA..
ejemplo: inversa(mat) devuelve la matriz inversa de mat donde mat es una matriz cuadrada de 2x2.

```python
"""El código que proporcionó crea una matriz de enteros de 2x2 y luego la imprime en la consola."""
"""la matrizvariable
El forciclo itera a través de las filas de la matriz (el ciclo externo)
Para cada elemento de la matriz, la printinstrucción se denomina endconjunto de parámetros con un carácter de espacio. Este
Después de imprimir cada fila de la matriz, printse vuelve a llamar a la declaración sin ningún argumento. """
# Creamos la matriz
matriz = [[1, 2], [3, 4]]
# Imprimimos la matriz
for i in range(2):
    for j in range(2):
        print(matriz[i][j], end=' ')
    print()
```

```python

# Función para calcular la inversa de una matriz de 2x2
def inversa_2x2(matriz):
    # Calculamos el determinante
    determinante = matriz[0][0] * matriz[1][1] - matriz[0][1] * matriz[1][0]

    # Comprobamos si el determinante es igual a cero
    if determinante == 0:
        print("La matriz no tiene inversa.")
        return None

    # Calculamos la matriz adjunta
    adjunta = [[matriz[1][1], -matriz[0][1]], [-matriz[1][0], matriz[0][0]]]

    # Dividimos la matriz adjunta por el determinante para obtener la inversa
    inversa = [[adjunta[0][0]/determinante, adjunta[0][1]/determinante],
               [adjunta[1][0]/determinante, adjunta[1][1]/determinante]]

    return inversa

# Ejemplo de uso
matriz = [[1, 2], [3, 4]]
inversa = inversa_2x2(matriz)
print("Matriz original:")
for fila in matriz:
    print(fila)
print("Inversa:")
for fila in inversa:
    print(fila)
```

```python
"""​ax^2 + bx + c = 0 por x1 y x2 utilizando la fórmula cuadrática

El código primero importa elmathbibliotecasqrtfunción
Los coeficientes dea = 1, b = -(10**10 + 10**-10)y c = 1.0.
Luego, el código calcula elx = (-b ± sqrt(b^2 - 4ac)) / 2a.
Los valores calculadosx1yx2sonprintfunción."""

import math
 #Importaremos la libreria necesaria para calculo de la raiz en la parte del discriminante
## Ponemos los coeficientes de la ecuación ax^2 + bx + c que nos han dado

a= 1 # coeficiente cuadratico (ax^2)
b= -(10**10 + 10**-10) # coeficiente lineal (bx)
c=1.0 # termino independiente (c)
x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a) #aplicamos la formula para x1,x2
x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)

print("x1 =", x1)   #Mostramos los datos obtenidos
print("x2 =", x2)
```

**¿Qué resultado se obtiene?** $x_{1}= \frac{10^{10}+10^{10}}{2} = 1*10^{10} = 10000000000.0$ y $x_{2} =\frac{10^{10}-10^{10}}{2} = \frac{0}{2} =  0.0$ \\


**¿se presenta cancelación numérica?** Podemos ver que hay una cancelación numérica enorme en el cálculo de $x_{2}$, ya que el segundo término dentro de la raíz cuadrada en la fórmula general, es muy cercano a b y luego se resta, lo que puede dar lugar a errores de redondeo.
En este caso, b es un número muy grande ($10^{10}$) y b es un número muy pequeño ($10^{-10}$) a la vez, por lo que la resta de ambos resulta en una pérdida significativa de precisión, que se manifiesta en la solución de $x_{2}$ que es muy pequeña en comparación con b

**¿por qué sucede la cancelación numérica?** Este tipo de cancelación numérica puede ocurrir en cálculos numéricos cuando se restan dos números que son muy cercanos entre sí, lo que puede conducir a una pérdida de precisión en el resultado final. Para evitar este tipo de cancelación numérica, se pueden utilizar técnicas de estabilidad numérica como la reorganización algebraica de las fórmulas para minimizar los errores de redondeo, o usar algoritmos más precisos que manejen los números con mayor precisión \\
