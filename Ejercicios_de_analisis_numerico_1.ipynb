Por lo tanto, la norma L2 cumple la tercera propiedad.

Como la norma L2 cumple las tres propiedades, podemos concluir que es una norma vectorial.

\section*{Demostración de desigualdades entre normas}

Para demostrar estas desigualdades, utilizaremos las definiciones de las normas vectoriales
$\|\cdot\|_{\infty}$ y $\|\cdot\|_{2}$.

La norma $\|\cdot\|_{\infty}$, también conocida como la norma del máximo, se define como:
\[
\|\vec{x}\|_{\infty} = \max_{1 \le i \le n} |x_i|.
\]

La norma $\|\cdot\|_{2}$, también conocida como la norma euclidiana, se define como:
\[
\|\vec{x}\|_{2} = \sqrt{\sum_{i=1}^{n} x_i^2}.
\]

\subsection*{Primera desigualdad}

Comencemos por la primera desigualdad:
\[
\begin{aligned}
\|\vec{x}\|_{\infty}
&= \max_{1 \le i \le n} |x_i| \\
&\le \sqrt{\sum_{i=1}^{n} x_i^2}
\qquad \text{(desigualdad de Cauchy--Schwarz)} \\
&= \|\vec{x}\|_{2}.
\end{aligned}
\]

\subsection*{Segunda desigualdad}

Para la segunda desigualdad, procedemos de la siguiente manera:
\[
\begin{aligned}
\|\vec{x}\|_{2}
&= \sqrt{\sum_{i=1}^{n} x_i^2} \\
&\le \sqrt{\sum_{i=1}^{n} \left( \max_{1 \le i \le n} |x_i| \right)^2}
\qquad \text{(cada término es menor o igual al máximo)} \\
&= \sqrt{n}\,\|\vec{x}\|_{\infty}.
\end{aligned}
\]

\subsection*{Conclusión}

Por lo tanto, hemos demostrado que:
\[
\|\vec{x}\|_{\infty}
\le
\|\vec{x}\|_{2}
\le
\sqrt{n}\,\|\vec{x}\|_{\infty},
\]
lo cual completa la demostración.

    $f(v) = 0$ si y solo si $v = 0$.
  </p>

  <p align="justify">

  </p>

    $f(kv) = f(0) = \|0\| = 0 = |k|\,f(v)$.
  </p>

  <p align="justify">
    Si $k > 0$, entonces
    $f(kv) = \|kv\| = k\|v\| = kf(v) = |k|\,f(v)$.
  </p>

  <p align="justify">
    Si $k < 0$, entonces
    $f(kv) = \|kv\| = |-k|\,\|v\| = |k|\,\|v\| = |k|\,f(v)$.
  </p>

  <p align="justify">
    Por lo tanto, en todos los casos se cumple que $f(kv) = |k|\,f(v)$.
  </p>

  <p align="justify">

    $\|u+v\| \le \|u\| + \|v\|$ para todo $u, v \in V$.

##1.   Norma L1 (también conocida como "norma de Manhattan" o "norma de taxista"):


<font color="Teal" face="Comic Sans MS,arial">
  <h3 align="center"><i>Demostración: Norma de Manhattan</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    Para demostrar que la norma de Manhattan es efectivamente una norma vectorial, primero
    recordemos su definición. La norma de Manhattan, también conocida como norma $L^1$,
    es una medida de distancia en un espacio euclidiano.
  </p>

  <p align="justify">
    En un espacio euclidiano de $n$ dimensiones, la norma de Manhattan de un vector
    $\vec{x} = (x_1, x_2, \dots, x_n)$ se define como:
  </p>

  <p align="center">
    $\|\vec{x}\|_1 = |x_1| + |x_2| + \cdots + |x_n|$
  </p>

  <p align="justify">
    A continuación, verificaremos que esta función cumple las tres propiedades que definen
    una norma vectorial.
  </p>

  <p align="justify">
    <b>1. Propiedad de positividad.</b><br>
    Demostraremos que $\|\vec{x}\|_1 \ge 0$ para todo vector $\vec{x}$, y que
    $\|\vec{x}\|_1 = 0$ si y solo si $\vec{x} = \vec{0}$.
  </p>

  <p align="justify">
    Cada término de la suma $|x_i|$ es no negativo, ya que es el valor absoluto de un número real.
    Por lo tanto, la suma de estos términos también es no negativa. Además,
    $\|\vec{x}\|_1 = 0$ únicamente cuando todos los componentes $x_i = 0$, es decir,
    cuando $\vec{x} = \vec{0}$.
  </p>

  <p align="justify">
    <b>2. Propiedad de homogeneidad.</b><br>
    Demostraremos que para todo escalar real $a$ y todo vector $\vec{x}$ se cumple:
    $\|\;a\vec{x}\;\|_1 = |a|\,\|\vec{x}\|_1$.
  </p>

  <p align="center">
    $\|a\vec{x}\|_1 = |ax_1| + |ax_2| + \cdots + |ax_n|$
  </p>

  <p align="center">
    $= |a||x_1| + |a||x_2| + \cdots + |a||x_n|$
  </p>

  <p align="center">
    $= |a|(|x_1| + |x_2| + \cdots + |x_n|) = |a|\,\|\vec{x}\|_1$
  </p>

  <p align="justify">
    Por lo tanto, la norma de Manhattan satisface la propiedad de homogeneidad.
  </p>

  <p align="justify">
    <b>3. Propiedad de desigualdad triangular.</b><br>
    Demostraremos que para cualesquiera vectores $\vec{x}$ y $\vec{y}$ se cumple:
    $\|\vec{x} + \vec{y}\|_1 \le \|\vec{x}\|_1 + \|\vec{y}\|_1$.
  </p>

  <p align="center">
    $\|\vec{x} + \vec{y}\|_1 = |x_1 + y_1| + |x_2 + y_2| + \cdots + |x_n + y_n|$
  </p>

  <p align="center">
    $\le |x_1| + |y_1| + |x_2| + |y_2| + \cdots + |x_n| + |y_n|$
  </p>

  <p align="center">
    $= \|\vec{x}\|_1 + \|\vec{y}\|_1$
  </p>

  <p align="justify">
    Por lo tanto, la norma de Manhattan satisface la desigualdad triangular.
  </p>

  <p align="justify">
    En conclusión, hemos demostrado que la norma de Manhattan cumple las propiedades de
    positividad, homogeneidad y desigualdad triangular, por lo que es una norma vectorial válida.
  </p>
</font>


>La norma L1 de un vector x de dimensión n se define como la suma de los valores absolutos de sus componentes. La función correspondiente es:

```python
def norm_L1(x):
    return sum(abs(x[i]) for i in range(len(x)))
```

<font color="Teal" face="Comic Sans MS,arial">
  <h3 align="center"><i>Recordatorio teórico (breve)</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    La distancia de Manhattan entre dos puntos $(x_1, y_1)$ y $(x_2, y_2)$ es:
  </p>

  <p align="center">
    $d_M = |x_1 - x_2| + |y_1 - y_2|$
  </p>

  <p align="justify">
    Se interpreta como el recorrido mínimo siguiendo calles ortogonales (como en Manhattan).
  </p>
</font>


```python
import math
import matplotlib.pyplot as plt


# =============================================================================
# Distancias en el plano
# =============================================================================

def manhattan_distance(x1, y1, x2, y2):
    """
    Calcula la distancia de Manhattan entre dos puntos en R².

    Parameters
    ----------
    x1, y1 : float
        Coordenadas del primer punto.
    x2, y2 : float
        Coordenadas del segundo punto.

    Returns
    -------
    float
        Distancia de Manhattan.
    """
    return abs(x1 - x2) + abs(y1 - y2)


def euclidean_distance(x1, y1, x2, y2):
    """
    Calcula la distancia Euclidiana entre dos puntos en R².
    """
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)


# =============================================================================
# Visualización
# =============================================================================

def plot_manhattan_vs_euclidean(x1, y1, x2, y2):
    """
    Grafica la distancia de Manhattan y la distancia Euclidiana
    entre dos puntos.
    """
    d_m = manhattan_distance(x1, y1, x2, y2)
    d_e = euclidean_distance(x1, y1, x2, y2)

    plt.figure(figsize=(7, 7))

    # Puntos
    plt.scatter([x1, x2], [y1, y2], color="black", zorder=5)
    plt.text(x1, y1, " A", fontsize=12)
    plt.text(x2, y2, " B", fontsize=12)

    # Trayectoria Manhattan (forma de L)
    plt.plot([x1, x2], [y1, y1], "r--", label="Manhattan (horizontal)")
    plt.plot([x2, x2], [y1, y2], "r--", label="Manhattan (vertical)")

    # Trayectoria Euclidiana (línea recta)
    plt.plot([x1, x2], [y1, y2], "b-", label="Euclidiana")

    plt.title("Distancia Manhattan vs Euclidiana")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(alpha=0.4)
    plt.axis("equal")

    plt.legend(
        title=f"d_M = {d_m:.2f}\n"
              f"d_E = {d_e:.2f}"
    )

    plt.show()


# =============================================================================
# Ejemplo de uso
# =============================================================================

if __name__ == "__main__":
    x1, y1 = 1, 2
    x2, y2 = 4, 6

    d = manhattan_distance(x1, y1, x2, y2)
    print(f"Distancia de Manhattan: {d}")

    plot_manhattan_vs_euclidean(x1, y1, x2, y2)
```

  <p align="justify">

  </p>

  <p align="center">
    $\|\vec{v}\|_2 = \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}$
  </p>

  <p align="justify">
    Por ejemplo, si tenemos el vector $(3,4)$, su norma L2 es:
  </p>

  <p align="center">
    $\|(3,4)\|_2 = \sqrt{3^2 + 4^2} = 5$
  </p>

  <p align="justify">
    La norma L2 tiene propiedades importantes, como la desigualdad triangular
    $\|\vec{u} + \vec{v}\|_2 \le \|\vec{u}\|_2 + \|\vec{v}\|_2$ y la invariancia bajo rotaciones.
    Además, la norma L2 se utiliza a menudo como función de pérdida en algoritmos de aprendizaje
    automático, como regresión lineal o regresión logística.
  </p>
</font>


<font color="Teal" face="Comic Sans MS,arial">
  <h3 align="center"><i>Demostración: Norma Euclidiana (Norma L2)</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    Para demostrar que la norma euclidiana es una norma vectorial, recordemos su definición.
    En un espacio euclidiano de $n$ dimensiones, la norma euclidiana (o norma $L^2$) de un vector
    $\vec{x} = (x_1, x_2, \dots, x_n)$ se define como:
  </p>

  <p align="center">
    $\|\vec{x}\|_2 = \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}$
  </p>

  <p align="justify">
    Verificaremos que esta función cumple las tres propiedades que definen una norma:
    positividad, homogeneidad y desigualdad triangular.
  </p>

  <p align="justify">
    <b>1. Propiedad de positividad.</b><br>
    Demostraremos que $\|\vec{x}\|_2 \ge 0$ para todo $\vec{x}$ y que $\|\vec{x}\|_2 = 0$
    si y solo si $\vec{x} = \vec{0}$.
  </p>

  <p align="justify">
    Cada término $x_i^2$ es no negativo por ser un cuadrado. Por tanto,
    $\sum_{i=1}^n x_i^2 \ge 0$ y su raíz cuadrada también es no negativa, es decir,
    $\|\vec{x}\|_2 \ge 0$. Además, $\|\vec{x}\|_2 = 0$ implica
    $\sum_{i=1}^n x_i^2 = 0$, lo cual sólo ocurre cuando $x_i = 0$ para todo $i$,
    es decir, cuando $\vec{x} = \vec{0}$.
  </p>

  <p align="justify">
    <b>2. Propiedad de homogeneidad.</b><br>
    Demostraremos que para todo escalar real $a$ se cumple $\|a\vec{x}\|_2 = |a|\,\|\vec{x}\|_2$.
  </p>

  <p align="center">
    $\|a\vec{x}\|_2
    = \sqrt{(ax_1)^2 + (ax_2)^2 + \cdots + (ax_n)^2}$
  </p>

  <p align="center">
    $= \sqrt{a^2(x_1^2 + x_2^2 + \cdots + x_n^2)}
    = |a|\,\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}
    = |a|\,\|\vec{x}\|_2$
  </p>

  <p align="justify">
    Por lo tanto, la norma euclidiana satisface la propiedad de homogeneidad.
  </p>

  <p align="justify">
    <b>3. Propiedad de desigualdad triangular.</b><br>
    Demostraremos que para cualesquiera vectores $\vec{x}$ y $\vec{y}$ se cumple:
    $\|\vec{x} + \vec{y}\|_2 \le \|\vec{x}\|_2 + \|\vec{y}\|_2$.
  </p>

  <p align="justify">
    Usamos la desigualdad de Cauchy–Schwarz, que establece que:
    $|\langle \vec{x}, \vec{y}\rangle| \le \|\vec{x}\|_2\,\|\vec{y}\|_2$,
    donde $\langle \vec{x}, \vec{y}\rangle$ es el producto interno.
  </p>

  <p align="center">
    $\|\vec{x} + \vec{y}\|_2^2
    = \langle \vec{x} + \vec{y}, \vec{x} + \vec{y}\rangle
    = \|\vec{x}\|_2^2 + 2\langle \vec{x}, \vec{y}\rangle + \|\vec{y}\|_2^2$
  </p>

  <p align="center">
    $\le \|\vec{x}\|_2^2 + 2\|\vec{x}\|_2\|\vec{y}\|_2 + \|\vec{y}\|_2^2
    = (\|\vec{x}\|_2 + \|\vec{y}\|_2)^2$
  </p>

  <p align="justify">
    Tomando raíz cuadrada en ambos lados (y usando que ambos lados son $\ge 0$), obtenemos:
    $\|\vec{x} + \vec{y}\|_2 \le \|\vec{x}\|_2 + \|\vec{y}\|_2$.
  </p>

  <p align="justify">
    En conclusión, la norma euclidiana cumple positividad, homogeneidad y desigualdad triangular,
    por lo que es una norma vectorial válida y una medida de longitud/distancia fundamental en
    matemáticas, física e ingeniería.
  </p>
</font>


>La norma L2 de un vector x de dimensión n se define como la raíz cuadrada de la suma de los cuadrados de sus componentes. La función correspondiente es:

```python
import math
import matplotlib.pyplot as plt


# =============================================================================
# Distancia Euclidiana
# =============================================================================

def euclidean_distance(x1, y1, x2, y2, decimals=None):
    """
    Calcula la distancia euclidiana entre dos puntos en R².

    Parameters
    ----------
    x1, y1 : float
        Coordenadas del primer punto.
    x2, y2 : float
        Coordenadas del segundo punto.
    decimals : int or None
        Número de decimales para redondear el resultado.
        Si es None, se devuelve el valor completo.

    Returns
    -------
    float
        Distancia euclidiana.
    """
    d = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    return round(d, decimals) if decimals is not None else d


# =============================================================================
# Visualización
# =============================================================================

def plot_euclidean_distance(x1, y1, x2, y2):
    """
    Grafica la distancia euclidiana entre dos puntos.
    """
    d = euclidean_distance(x1, y1, x2, y2)

    plt.figure(figsize=(7, 7))

    # Puntos
    plt.scatter([x1, x2], [y1, y2], color="black", zorder=5)
    plt.text(x1, y1, " A", fontsize=12)
    plt.text(x2, y2, " B", fontsize=12)

    # Segmento euclidiano
    plt.plot([x1, x2], [y1, y2], "b-", lw=2, label="Distancia Euclidiana")

    # Proyecciones para ver el triángulo rectángulo
    plt.plot([x1, x2], [y1, y1], "r--", alpha=0.6)
    plt.plot([x2, x2], [y1, y2], "r--", alpha=0.6)

    plt.title(f"Distancia Euclidiana  d = {d:.4f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(alpha=0.4)
    plt.axis("equal")
    plt.legend()
    plt.show()


# =============================================================================
# Ejemplo de uso
# =============================================================================

if __name__ == "__main__":
    x1, y1 = 1, 2
    x2, y2 = 4, 6

    print("Distancia exacta (float):", euclidean_distance(x1, y1, x2, y2))
    print("Distancia redondeada a 2 decimales:", euclidean_distance(x1, y1, x2, y2, decimals=2))

    plot_euclidean_distance(x1, y1, x2, y2)
```

##3.-Norma infinito (también conocida como "norma máxima" o "norma de Chebyshev"):


La norma máxima (también conocida como norma infinito o norma de Chebyshev) es una medida de distancia entre dos puntos en un espacio métrico. En un plano cartesiano bidimensional, la norma máxima de dos puntos (x1, y1) y (x2, y2) se define como la máxima diferencia entre sus coordenadas

<font color="Teal" face="Comic Sans MS,arial">
  <h3 align="center"><i>Demostración: Norma Máxima (Norma Infinita / Chebyshev)</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    La norma máxima, también conocida como norma infinita o norma de Chebyshev, se define para un
    vector $\vec{x}=(x_1,x_2,\dots,x_n)$ como:
  </p>

  <p align="center">
    $\|\vec{x}\|_{\infty} = \max\left(|x_1|,|x_2|,\dots,|x_n|\right)$
  </p>

  <p align="justify">
    Para demostrar que esta función es una norma, verificaremos que cumple las tres propiedades:
    positividad, homogeneidad y desigualdad triangular.
  </p>

  <p align="justify">
    <b>1. Propiedad de positividad.</b><br>
    Demostraremos que $\|\vec{x}\|_{\infty} \ge 0$ para todo $\vec{x}$ y que
    $\|\vec{x}\|_{\infty}=0$ si y solo si $\vec{x}=\vec{0}$.
  </p>

  <p align="justify">
    Como $|x_i|\ge 0$ para todo $i$, el máximo de estos valores también es no negativo, por lo que
    $\|\vec{x}\|_{\infty}\ge 0$. Además, $\|\vec{x}\|_{\infty}=0$ implica que
    $\max(|x_1|,\dots,|x_n|)=0$, lo cual sólo es posible si $|x_i|=0$ para todo $i$, es decir,
    $x_i=0$ para todo $i$ y por tanto $\vec{x}=\vec{0}$.
  </p>

  <p align="justify">
    <b>2. Propiedad de homogeneidad.</b><br>
    Demostraremos que para todo escalar real $a$ se cumple:
    $\|a\vec{x}\|_{\infty} = |a|\,\|\vec{x}\|_{\infty}$.
  </p>

  <p align="center">
    $\|a\vec{x}\|_{\infty}
    = \max\left(|ax_1|,|ax_2|,\dots,|ax_n|\right)
    = \max\left(|a||x_1|,|a||x_2|,\dots,|a||x_n|\right)$
  </p>

  <p align="center">
    $= |a|\max\left(|x_1|,|x_2|,\dots,|x_n|\right)
    = |a|\,\|\vec{x}\|_{\infty}$
  </p>

  <p align="justify">
    Por lo tanto, la norma máxima satisface la propiedad de homogeneidad.
  </p>

  <p align="justify">
    <b>3. Propiedad de desigualdad triangular.</b><br>
    Demostraremos que para cualesquiera vectores $\vec{x}$ y $\vec{y}$ se cumple:
    $\|\vec{x}+\vec{y}\|_{\infty} \le \|\vec{x}\|_{\infty} + \|\vec{y}\|_{\infty}$.
  </p>

  <p align="justify">
    Para cada componente $i$, por la desigualdad triangular en $\mathbb{R}$ se tiene:
    $|x_i+y_i| \le |x_i| + |y_i|$.
    Además, como $|x_i|\le \|\vec{x}\|_{\infty}$ y $|y_i|\le \|\vec{y}\|_{\infty}$, entonces:
  </p>

  <p align="center">
    $|x_i+y_i| \le |x_i| + |y_i| \le \|\vec{x}\|_{\infty} + \|\vec{y}\|_{\infty}$
  </p>

  <p align="justify">
    Tomando el máximo sobre $i=1,\dots,n$, obtenemos:
  </p>

  <p align="center">
    $\|\vec{x}+\vec{y}\|_{\infty}
    = \max_{1\le i\le n}|x_i+y_i|
    \le \|\vec{x}\|_{\infty} + \|\vec{y}\|_{\infty}$
  </p>

  <p align="justify">
    En conclusión, la norma máxima cumple positividad, homogeneidad y desigualdad triangular,
    por lo que es una norma vectorial válida.
  </p>
</font>

## NORMAS INDUCIDAS

2.1 (0.75 puntos) Sea $A,B\in M_{n\times n}$ sobre $\mathbb{R} , \left\Vert \cdotp\right\Vert$  una norma vectorial cualquiera y la definición $\left\Vert A\right\Vert = \underset{\vec{x}\neq\vec{0}}{sup}\frac{\left\Vert A\vec{x}\right\Vert }{\left\Vert \vec{x}\right\Vert }$ demostrar que:

1.   $\left\Vert A\right\Vert \geq0$.
2.   $\left\Vert A\right\Vert =0\,\Longleftrightarrow A=0$.
3.   $\left\Vert \alpha A\right\Vert =\left|\alpha\right|\left\Vert A\right\Vert$.
4.   $\left\Vert A+B\right\Vert \leq\left\Vert A\right\Vert +\left\Vert B\right\Vert$.



#### Demostración

1.- Por definición, $\left\Vert A\right\Vert = \underset{\vec{x}\neq\vec{0}} {sup}\frac{\left\Vert A\vec{x}\right\Vert }{\left\Vert \vec{x}\right\Vert }$
por lo que el valor de $\left\Vert A\right\Vert$ siempre será mayor o igual a cero.
Esto se debe a que tanto el numerador como el denominador de la fracción son valores no negativos. Por lo tanto, se cumple que $\left\Vert A\right\Vert \geq0$.

2.- Si $\left\Vert A\right\Vert =0$, entonces el valor del supremo es cero. Esto significa que $\frac{\left\Vert A\vec{x}\right\Vert }{\left\Vert \vec{x}\right\Vert } = 0$ para todo $\vec{x}\neq \vec{0}$. Pero esto solo es posible si $A\vec{x}=\vec{0}$ para todo $\vec{x}\neq \vec{0}$, lo que implica que $A=0$.

Por otro lado, si $A=0$, entonces $\left\Vert A\vec{x}\right\Vert =0$ para todo $\vec{x}\in \mathbb{R}^n$, por lo que $\frac{\left\Vert A\vec{x}\right\Vert }{\left\Vert \vec{x}\right\Vert }=0$ para todo $\vec{x}\neq \vec{0}$. Por lo tanto, $\left\Vert A\right\Vert =0$.

3.-
Para demostrar que $\left\Vert \alpha A\right\Vert =\left|\alpha\right|\left\Vert A\right\Vert$, se procede de la siguiente manera:

\begin{align*}
\left\Vert \alpha A\right\Vert &= \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert \alpha A\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \sup_{\vec{x}\neq\vec{0}} \frac{\left|\alpha\right|\left\Vert A\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \left|\alpha\right|\sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert A\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \left|\alpha\right|\left\Vert A\right\Vert
\end{align*}

Por lo tanto, se ha demostrado que $\left\Vert \alpha A\right\Vert =\left|\alpha\right|\left\Vert A\right\Vert$.

4.- Para demostrar que $\left\Vert A+B\right\Vert \leq\left\Vert A\right\Vert +\left\Vert B\right\Vert$, se procede de la siguiente manera:

\begin{align*}
\left\Vert A+B\right\Vert &= \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert (A+B)\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert A\vec{x}+B\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&\leq \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert A\vec{x}\right\Vert+\left\Vert B\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert A\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert} + \sup_{\vec{x}\neq\vec{0}} \frac{\left\Vert B\vec{x}\right\Vert}{\left\Vert \vec{x}\right\Vert}\
&= \left\Vert A\right\Vert + \left\Vert B\right\Vert
\end{align*}
Por lo tanto, se ha demostrado que $\left\Vert A+B\right\Vert \leq\left\Vert A\right\Vert +\left\Vert B\right\Vert$.

  <h3 align="center"><i>Norma matricial</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    <b>Demostración de las propiedades de norma para la norma matricial.</b>
    Sea $\|A\|$ la norma matricial inducida por una norma vectorial, definida como:
  </p>

  <p align="center">
    $\|A\| = \sup_{\vec{x}\neq \vec{0}} \dfrac{\|A\vec{x}\|}{\|\vec{x}\|}$
  </p>

  <p align="justify">
    <b>1) No negatividad:</b> $\|A\|\ge 0$.<br>
    Por definición, $\|A\|$ es el supremo de $\dfrac{\|A\vec{x}\|}{\|\vec{x}\|}$ para todo
    $\vec{x}\neq \vec{0}$. Como la norma vectorial es siempre no negativa, se tiene que
    $\|A\vec{x}\|\ge 0$ y $\|\vec{x}\|>0$ para $\vec{x}\neq \vec{0}$, por lo tanto:
  </p>

  <p align="center">
    $\dfrac{\|A\vec{x}\|}{\|\vec{x}\|}\ge 0 \quad \text{para todo } \vec{x}\neq \vec{0}.$
  </p>

  <p align="justify">
    El supremo de un conjunto de valores no negativos es no negativo; por consiguiente,
    $\|A\|\ge 0$.
  </p>

  <p align="justify">
    <b>2) Homogeneidad:</b> $\|\alpha A\| = |\alpha|\,\|A\|$.<br>
    Consideremos $\vec{x}\neq \vec{0}$. Entonces, usando la propiedad $\|\alpha \vec{z}\|=|\alpha|\|\vec{z}\|$:
  </p>

  <p align="center">
    $\|\alpha A\|
    = \sup_{\vec{y}\neq \vec{0}} \dfrac{\|\alpha A\vec{y}\|}{\|\vec{y}\|}
    = \sup_{\vec{y}\neq \vec{0}} \dfrac{|\alpha|\|A\vec{y}\|}{\|\vec{y}\|}
    = |\alpha|\sup_{\vec{y}\neq \vec{0}} \dfrac{\|A\vec{y}\|}{\|\vec{y}\|}
    = |\alpha|\,\|A\|.$
  </p>

  <p align="justify">
    Por lo tanto, $\|\alpha A\| = |\alpha|\,\|A\|$.
  </p>

  <p align="justify">
    <b>3) Desigualdad triangular:</b> $\|A+B\| \le \|A\| + \|B\|$.<br>
    Para cualquier $\vec{x}\neq \vec{0}$, por la desigualdad triangular de la norma vectorial:
  </p>

  <p align="center">
    $\|(A+B)\vec{x}\|
    = \|A\vec{x}+B\vec{x}\|
    \le \|A\vec{x}\| + \|B\vec{x}\|.$
  </p>

  <p align="justify">
    Además, por definición de norma inducida se cumple $\|A\vec{x}\|\le \|A\|\|\vec{x}\|$ y
    $\|B\vec{x}\|\le \|B\|\|\vec{x}\|$. Sustituyendo:
  </p>

  <p align="center">
    $\|(A+B)\vec{x}\|
    \le \|A\|\,\|\vec{x}\| + \|B\|\,\|\vec{x}\|
    = (\|A\|+\|B\|)\,\|\vec{x}\|.$
  </p>

  <p align="justify">
    Dividiendo entre $\|\vec{x}\|>0$:
  </p>

  <p align="center">
    $\dfrac{\|(A+B)\vec{x}\|}{\|\vec{x}\|}
    \le \|A\|+\|B\| \quad \text{para todo } \vec{x}\neq \vec{0}.$
  </p>

  <p align="justify">
    Tomando supremo en $\vec{x}\neq \vec{0}$, obtenemos:
  </p>

  <p align="center">
    $\|A+B\| \le \|A\| + \|B\|.$
  </p>
</font>


```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# Normas matriciales: norma 1 y norma infinito
# =============================================================================

# Matriz de ejemplo
A = np.array([[1, 2],
              [12, 3]])

# -----------------------------
# Norma 1 (máxima suma por columnas)
# -----------------------------
def norma_uno(A: np.ndarray) -> float:
    """
    Calcula la norma 1 de una matriz.

    ||A||_1 = max_j sum_i |a_ij|
    """
    columnas = np.sum(np.abs(A), axis=0)
    return np.max(columnas)


# -----------------------------
# Norma infinito (máxima suma por filas)
# -----------------------------
def norma_infinito(A: np.ndarray) -> float:
    """
    Calcula la norma infinito de una matriz.

    ||A||_∞ = max_i sum_j |a_ij|
    """
    filas = np.sum(np.abs(A), axis=1)
    return np.max(filas)


# -----------------------------
# Resultados numéricos
# -----------------------------
norma1 = norma_uno(A)
normaInf = norma_infinito(A)

print("Matriz A:\n", A)
print("\nNorma 1 (implementación propia):", norma1)
print("Norma ∞ (implementación propia):", normaInf)

print("\nNorma 1 (NumPy):", np.linalg.norm(A, 1))
print("Norma ∞ (NumPy):", np.linalg.norm(A, np.inf))


# =============================================================================
# Visualización
# =============================================================================

def plot_normas(A: np.ndarray):
    """
    Grafica las sumas por columnas y por filas para ilustrar
    la norma 1 y la norma infinito.
    """
    columnas = np.sum(np.abs(A), axis=0)
    filas = np.sum(np.abs(A), axis=1)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # Gráfica de columnas (norma 1)
    axes[0].bar(range(1, len(columnas)+1), columnas)
    axes[0].set_title("Suma de valores absolutos por columna")
    axes[0].set_xlabel("Columna j")
    axes[0].set_ylabel(r"$\sum_i |a_{ij}|$")
    axes[0].axhline(norma1, color="red", linestyle="--", label="Norma 1")
    axes[0].legend()
    axes[0].grid(alpha=0.3)

    # Gráfica de filas (norma infinito)
    axes[1].bar(range(1, len(filas)+1), filas)
    axes[1].set_title("Suma de valores absolutos por fila")
    axes[1].set_xlabel("Fila i")
    axes[1].set_ylabel(r"$\sum_j |a_{ij}|$")
    axes[1].axhline(normaInf, color="red", linestyle="--", label="Norma ∞")
    axes[1].legend()
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.show()


# Mostrar gráficas
plot_normas(A)
```

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    <b>Demostración de las propiedades de norma para la norma espectral.</b><br>
    Para la norma espectral, consideramos la definición:
  </p>

  <p align="center">
    $\|A\| = \sqrt{\rho(A^T A)}$
  </p>

  <p align="justify">
    donde $\rho(\cdot)$ denota el radio espectral de una matriz. A partir de esta definición,
    demostramos las propiedades de norma:
  </p>

  <p align="justify">
    <b>1) No negatividad:</b> $\|A\|\ge 0$.<br>
    Como $\rho(\cdot)$ es no negativo por definición y la raíz cuadrada preserva no negatividad,
    se tiene:
  </p>

  <p align="center">
    $\|A\| = \sqrt{\rho(A^T A)} \ge 0.$
  </p>

  <p align="justify">
    <b>2) Definitud:</b> $\|A\|=0 \iff A=0$.<br>
    Si $\|A\|=0$, entonces $\rho(A^T A)=0$. Dado que $A^T A$ es semidefinida positiva,
    $\rho(A^T A)=0$ implica que todos sus autovalores son $0$, por lo tanto $A^T A=0$ y se sigue
    que $A=0$. Recíprocamente, si $A=0$, entonces $\|A\|=0$.
  </p>

  <p align="justify">
    <b>3) Homogeneidad:</b> $\|\alpha A\| = |\alpha|\,\|A\|$ para todo escalar $\alpha$.<br>
    En efecto:
  </p>

  <p align="center">
    $
    \begin{aligned}
    \|\alpha A\|
    &= \sqrt{\rho\big((\alpha A)^T(\alpha A)\big)} \\
    &= \sqrt{\rho\big(\alpha^2 A^T A\big)} \\
    &= \sqrt{\alpha^2\,\rho(A^T A)} \\
    &= |\alpha|\,\sqrt{\rho(A^T A)} \\
    &= |\alpha|\,\|A\|.
    \end{aligned}
    $
  </p>

  <p align="justify">
    <b>4) Desigualdad triangular:</b> $\|A+B\|\le \|A\|+\|B\|$.<br>
    Para matrices $A$ y $B$, se tiene:
  </p>

  <p align="center">
    $
    \begin{aligned}
    \|A+B\|^2
    &= \rho\big((A+B)^T(A+B)\big) \\
    &= \rho\big(A^TA + A^TB + B^TA + B^TB\big) \\
    &\le \rho(A^TA) + 2\,\rho(A^TB) + \rho(B^TB) \\
    &= \|A\|^2 + 2\sqrt{\rho(A^TA)\rho(B^TB)} + \|B\|^2 \\
    &\le \|A\|^2 + 2\|A\|\,\|B\| + \|B\|^2 \\
    &= (\|A\|+\|B\|)^2.
    \end{aligned}
    $
  </p>

  <p align="justify">
    Tomando raíz cuadrada en ambos lados, obtenemos:
  </p>

  <p align="center">
    $\|A+B\| \le \|A\| + \|B\|,$
  </p>

  <p align="justify">
    como se quería demostrar.
  </p>

### Norma de Frobenius

Demostracion de las propiedades de norma para la norma de Frobenius:

1.- $\left\Vert A\right\Vert_F=\sqrt{\sum_{i=1}^n \sum_{j=1}^m |a_{ij}|^2}\geq0$
Claramente, cada término dentro de la raíz cuadrada es no negativo, ya que $|a_{ij}|\geq 0$ para todo $i$ y $j$, por lo que la suma de términos no negativos también es no negativa. Así, $\left\Vert A\right\Vert_F\geq0$.

2.- $\left\Vert A\right\Vert_F=0,\Longleftrightarrow A=0$
Si $A=0$, entonces $a_{ij}=0$ para todo $i$ y $j$, lo que implica que $\sum_{i=1}^n \sum_{j=1}^m |a_{ij}|^2=0$ y por lo tanto $\left\Vert A\right\Vert_F=0$.

Por otro lado, si $\left\Vert A\right\Vert_F=0$, entonces $\sum_{i=1}^n \sum_{j=1}^m |a_{ij}|^2=0$, lo que significa que $|a_{ij}|=0$ para todo $i$ y $j$. En otras palabras, $a_{ij}=0$ para todo $i$ y $j$, lo que implica que $A=0$.

3.- Para demostrar que $\left\Vert \alpha A\right\Vert =\left|\alpha\right|\left\Vert A\right\Vert$, primero notemos que la norma de Frobenius se define como $\left\Vert A\right\Vert_F = \sqrt{\sum_{i=1}^n\sum_{j=1}^m|a_{ij}|^2}$, donde $A\in M_{n\times m}$. Entonces, usando la definición de norma de Frobenius, tenemos:
\begin{align*}
\left|\alpha A\right|F &= \sqrt{\sum{i=1}^n\sum_{j=1}^m|\alpha a_{ij}|^2}\
&= \sqrt{\sum_{i=1}^n\sum_{j=1}^m|\alpha|^2|a_{ij}|^2}\
&= |\alpha|\sqrt{\sum_{i=1}^n\sum_{j=1}^m|a_{ij}|^2}\
&= |\alpha|\left|A\right|_F
\end{align*}

Por lo tanto, $\left\Vert \alpha A\right\Vert =\left|\alpha\right|\left\Vert A\right\Vert$.

4.-Para demostrar que $\left\Vert A+B\right\Vert_F \leq\left\Vert A\right\Vert_F +\left\Vert B\right\Vert_F$, primero notemos que:
Para ello, notemos que:

\begin{equation*}
\left(\sum_{i=1}^{n}\sum_{j=1}^{m} \left|a_{ij}+b_{ij}\right|^2\right)^{\frac{1}{2}} \leq \left(\sum_{i=1}^{n}\sum_{j=1}^{m} \left|a_{ij}\right|^2\right)^{\frac{1}{2}}+\left(\sum_{i=1}^{n}\sum_{j=1}^{m} \left|b_{ij}\right|^2\right)^{\frac{1}{2}}
\end{equation*}
Esta desigualdad se sigue directamente de la desigualdad triangular y el hecho de que la raíz cuadrada es una función creciente. Luego, usando la definición de norma de Frobenius, podemos reescribir la desigualdad anterior como:

$\left\Vert A+B\right\Vert_{F} \leq \left\Vert A\right\Vert_{F} + \left\Vert B\right\Vert_{F}$

Por lo tanto, $\left\Vert A+B\right\Vert_F \leq\left\Vert A\right\Vert_F +\left\Vert B\right\Vert_F$.


```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# Norma de Frobenius
# =============================================================================

# Matrices de ejemplo
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 3],
              [23, 34]])

print("Matriz A:\n", A)
print("\nMatriz B:\n", B)


# -----------------------------
# Implementación manual
# -----------------------------
def norma_frobenius(A: np.ndarray) -> float:
    """
    Calcula la norma de Frobenius de una matriz.

    ||A||_F = sqrt( sum_{i,j} a_ij^2 )
    """
    return np.sqrt(np.sum(A**2))


# Resultados
normA_manual = norma_frobenius(A)
normB_manual = norma_frobenius(B)

print("\nNorma de Frobenius (manual):")
print(f"||A||_F = {normA_manual:.4f}")
print(f"||B||_F = {normB_manual:.4f}")

print("\nNorma de Frobenius (NumPy):")
print(f"||A||_F = {np.linalg.norm(A, 'fro'):.4f}")
print(f"||B||_F = {np.linalg.norm(B, 'fro'):.4f}")


# =============================================================================
# Visualización
# =============================================================================

def plot_frobenius_components(A: np.ndarray, name="A"):
    """
    Grafica los cuadrados de los elementos de la matriz y
    muestra cómo contribuyen a la norma de Frobenius.
    """
    squared = A**2

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    # Matriz original
    im0 = axes[0].imshow(A, cmap="coolwarm")
    axes[0].set_title(f"Matriz {name}")
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            axes[0].text(j, i, A[i, j], ha="center", va="center", color="black")
    plt.colorbar(im0, ax=axes[0], fraction=0.046)

    # Cuadrados de los elementos
    im1 = axes[1].imshow(squared, cmap="viridis")
    axes[1].set_title(f"Cuadrados de {name}")
    for i in range(squared.shape[0]):
        for j in range(squared.shape[1]):
            axes[1].text(j, i, squared[i, j], ha="center", va="center", color="white")
    plt.colorbar(im1, ax=axes[1], fraction=0.046)

    plt.suptitle(
        f"||{name}||_F = sqrt(sum(a_ij^2)) = {norma_frobenius(A):.4f}",
        fontsize=12
    )
    plt.tight_layout()
    plt.show()


# Mostrar gráficos
plot_frobenius_components(A, name="A")
plot_frobenius_components(B, name="B")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# Normas vectoriales
# =============================================================================

def norma_euclidiana(vector):
    """
    Calcula la norma euclidiana (norma 2) de un vector.

    ||x||_2 = sqrt(sum_i x_i^2)
    """
    norm = 0.0
    for i in vector:
        norm += abs(i)**2
    return norm**0.5


def norma_p(vector, p=2):
    """
    Calcula la norma p de un vector usando NumPy.
    """
    return np.linalg.norm(vector, ord=p)


# =============================================================================
# Ejemplo
# =============================================================================

vec = np.array([1, 2, 3])

norm_manual = norma_euclidiana(vec)
norm_numpy = np.linalg.norm(vec, 2)

print("Vector:", vec)
print(f"Norma euclidiana (manual): {norm_manual:.4f}")
print(f"Norma euclidiana (NumPy):  {norm_numpy:.4f}")
print(f"Norma 1 (NumPy):           {np.linalg.norm(vec, 1):.4f}")
print(f"Norma infinito (NumPy):    {np.linalg.norm(vec, np.inf):.4f}")


# =============================================================================
# Visualización
# =============================================================================

def plot_vector_and_norm(vector):
    """
    Grafica el vector y muestra su longitud (norma euclidiana).
    """
    vector = np.array(vector)
    norm = np.linalg.norm(vector, 2)

    plt.figure(figsize=(6, 6))

    # Vector desde el origen
    plt.quiver(0, 0, vector[0], vector[1], angles='xy',
               scale_units='xy', scale=1, color="blue")

    # Proyecciones
    plt.plot([0, vector[0]], [0, 0], "r--", alpha=0.6)
    plt.plot([vector[0], vector[0]], [0, vector[1]], "r--", alpha=0.6)

    plt.text(vector[0]/2, vector[1]/2,
             f"||x||₂ = {norm:.2f}", fontsize=12)

    plt.xlim(-1, max(4, vector[0] + 1))
    plt.ylim(-1, max(4, vector[1] + 1))
    plt.xlabel("x₁")
    plt.ylabel("x₂")
    plt.title("Norma Euclidiana de un Vector (proyección 2D)")
    plt.grid(alpha=0.4)
    plt.axis("equal")
    plt.show()


# Gráfica solo para los primeros dos componentes
plot_vector_and_norm(vec[:2])
```

  <h3 align="center"><i>Norma vectorial</i></h3>
</font>

<font color="Black" face="Comic Sans MS,arial">
  <p align="justify">
    Una norma vectorial es una función que asigna un número no negativo a un vector en un
    espacio vectorial. Formalmente, una norma en un espacio vectorial <i>V</i> es una función
    $\|\cdot\| : V \rightarrow \mathbb{R}$ que satisface las siguientes propiedades:
  </p>

  <ul>
    <li>
      $\|u\| \ge 0$ para todo vector $u \in V$, y $\|u\| = 0$ si y solo si $u = \vec{0}$
      (donde $\vec{0}$ representa el vector nulo).
    </li>
    <li>
      $\|ku\| = |k|\,\|u\|$ para todo vector $u \in V$ y todo escalar $k$.
    </li>
    <li>
      $\|u + v\| \le \|u\| + \|v\|$ para todo vector $u, v \in V$
      (desigualdad del triángulo).
    </li>
  </ul>

  <p align="justify">
    Entonces, dada una norma vectorial $\|\cdot\|$, podemos definir la función
    $f : V \rightarrow \mathbb{R}$ que toma un vector $v$ y devuelve su norma, es decir,
    $f(v) = \|v\|$.
  </p>

  <p align="justify">
    Por lo tanto, cada norma vectorial define una función que toma como parámetro un vector
    y devuelve su respectiva norma.
  </p>
</font>


Para poder afirmar que una función sobre un espacio vectorial es norma, es necesario probar que cumple con la definición.

*P.D. 1-norma, es norma vectorial.*

*1.-*$$\left\Vert \vec{x}\right\Vert _{1}=\left|x_{1}\right|+\left|x_{2}\right|+\cdots+\left|x_{n}\right|\overset{\underbrace{prop.\,val.\,abs.}}{\geq}0$$

*2.-*$$\left\Vert \vec{x}\right\Vert _{1}=0\Longleftrightarrow \sum_{i=1}^{n}\left|x_{i}\right|=0\Longleftrightarrow x_{1}=x_{2}=\cdots=x_{n}=0\Longleftrightarrow\vec{x}=\vec{0}$$

*3.-*$$\begin{eqnarray*}
\left\Vert \alpha\vec{x}\right\Vert _{1} & \overset{\underbrace{def.}}{=} & \left|\alpha x_{1}\right|+\left|\alpha x_{2}\right|+\cdots+\left|\alpha x_{n}\right|\\
 & \overset{\underbrace{v.a.}}{=} & \left|\alpha\right|\left(\left|x_{1}\right|+\left|x_{2}\right|+\cdots+\left|x_{n}\right|\right)\\
 & = & \left|\alpha\right|\left\Vert \vec{x}\right\Vert _{1}
\end{eqnarray*}$$

*4.-*$$\begin{eqnarray*}
\left\Vert \vec{x}+\vec{y}\right\Vert _{1} & \overset{\underbrace{def.}}{=} & \sum_{i=1}^{n}\left|x_{i}+y_{i}\right|\overset{\underbrace{des.triang.}}{\leq}\sum_{i=1}^{n}\left|x_{i}\right|+\sum_{i=1}^{n}\left|y_{i}\right|\\
 & = & \left\Vert \vec{x}\right\Vert _{1}+\left\Vert \vec{y}\right\Vert _{1}
\end{eqnarray*}$$
