>a) Utiliza el metodo de Newton o el metodo de Lagrange para encontrar un polinomio P que interpole los datos; de tal manera que y_i = P(x_i), donde x_i representa los valores en el i-esimo periodo e y_i es el numero de personas activas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Cargar el archivo CSV
df = pd.read_csv('economia_poblacional.csv')

# Extraer las columnas de tiempo (xi) y personas activas (yi)
xi = df['fecha']  # Asegúrate de usar el nombre correcto de la columna del tiempo
yi = df['poblacion']  # Asegúrate de usar el nombre correcto de la columna de personas activas

# Método de Lagrange para encontrar el polinomio interpolante
poly = lagrange(xi, yi)

# Crear un rango de valores para el eje x
x_range = np.linspace(min(xi), max(xi), 100)

# Calcular los valores correspondientes en el eje y utilizando el polinomio interpolante
y_interp = poly(x_range)

# Graficar los datos y el polinomio interpolante
plt.plot(xi, yi, 'o', label='Datos reales')
plt.plot(x_range, y_interp, label='Polinomio interpolante')
plt.xlabel('fecha')
plt.ylabel('PIB')
plt.title('Interpolación del Producto Interno Bruto México')
plt.legend()
plt.show()
```

> c) Ahora encuentra polinomios Q_n(x), de grados n = 0, 1, 2, 5, 10, que ajusten los datos de la poblacion econ ́omicamente activa. Calcula el promedio de los residuales de cada ajuste yresponde ¿cual es el mejor ajuste?

> d) En un mismo plano grafica a los polinomios Q_n y los datos del archivo.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('economia_poblacional.csv')

# Extraer las columnas de tiempo (xi) y personas activas (yi)
xi = df['fecha']
yi = df['poblacion']

# Grados de los polinomios a ajustar
grados = [0, 1, 2, 5, 10]

# Inicializar una lista para almacenar los promedios de residuales
promedios_residuales = []

# Iterar sobre los diferentes grados
for grado in grados:
    # Ajustar el polinomio de grado n
    coeficientes = np.polyfit(xi, yi, grado)

    # Calcular los valores ajustados (polinomio evaluado en xi)
    yi_ajustado = np.polyval(coeficientes, xi)

    # Calcular los residuales
    residuales = yi - yi_ajustado

    # Calcular el promedio de los residuales
    promedio_residual = np.mean(np.abs(residuales))

    # Almacenar el promedio de residuales en la lista
    promedios_residuales.append(promedio_residual)

    # Graficar los datos y el polinomio ajustado
    plt.plot(xi, yi, 'o', label='Datos reales')
    plt.plot(xi, yi_ajustado, label=f'Polinomio Grado {grado}')
    plt.xlabel('fecha')
    plt.ylabel('poblacion')
    plt.title(f'Ajuste Polinómico de Grado {grado}')
    plt.legend()
    plt.show()

# Imprimir los promedios de residuales
for i, grado in enumerate(grados):
    print(f'Grado {grado}: Promedio de Residuales = {promedios_residuales[i]}')
```

e) Si tuvieramos la necesidad de pronosticar la cantidad de personas economicamente activas para el segundo trimestre de 2024, ¿qu ́e opci ́on ser ́ıa mejor: el polinomio interpolante de a) o algun polinomio de ajuste de c)?


La elección entre el polinomio interpolante obtenido en el paso a) y los polinomios de ajuste obtenidos en el paso c) dependerá de diversos factores. Aquí hay algunas consideraciones:

>Polinomio Interpolante (a):

Ventaja: Este polinomio pasa exactamente a través de todos los puntos de datos proporcionados.
Desventaja: Puede sufrir de oscilaciones no deseadas, especialmente si el grado del polinomio es alto. Este fenómeno se conoce como el fenómeno de Runge.

>Polinomios de Ajuste (c):

Ventaja: Los polinomios de ajuste pueden ser más estables y generalmente tienen menos oscilaciones excesivas.
Desventaja: No necesariamente pasan a través de todos los puntos de datos. El ajuste se realiza de manera más suave.
En la práctica, es común que los polinomios de ajuste sean preferidos para la extrapolación (pronóstico más allá de los datos conocidos) en lugar de los polinomios interpolantes. Los polinomios de ajuste tienden a ser más estables en regiones donde no hay datos de entrada.

Dicho esto, es importante tener en cuenta que todos los modelos polinómicos tienen limitaciones, especialmente cuando se extrapolan más allá del rango de los datos de entrenamiento. En casos donde la extrapolación es crítica, se pueden considerar otros métodos más avanzados, como modelos de regresión no polinómicos, modelos lineales, o incluso métodos basados en series temporales, dependiendo de la naturaleza de los datos y el problema.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
pobmex = pd.read_csv('poblacion-mexico.csv', thousands=',')
años = pobmex['Periodo']
poblacion = pobmex['Número de personas']

# Ajuste de polinomios
grados = [1, 3, 7, 10]

# Crear subgráficos
fig, axs = plt.subplots(len(grados), 1, figsize=(10, 6 * len(grados)))

for i, grado in enumerate(grados):
    # Ajustar un polinomio de grado n
    coeficientes = np.polyfit(años, poblacion, grado)

    # Crear un polinomio usando los coeficientes
    polinomio = np.poly1d(coeficientes)

    # Generar valores predichos para el polinomio
    valores_predichos = polinomio(años)

    # Calcular el residuo
    residuo = poblacion - valores_predichos

    # Calcular el promedio de los residuos
    promedio_residuos = np.mean(residuo)

    # Graficar el polinomio ajustado en un subgráfico
    axs[i].scatter(años, poblacion, label='Datos reales', color='blue', marker='o')
    axs[i].plot(años, valores_predichos, label=f'Grado {grado}', color='red')
    axs[i].set_title(f'Polinomio de grado {grado}')
    axs[i].set_xlabel('Periodo')
    axs[i].set_ylabel('Número de personas')
    axs[i].legend()
    axs[i].grid(True)

    # Imprimir el promedio de los residuos
    axs[i].text(0.05, 0.85, f'Promedio de residuos: {promedio_residuos:.2f}', transform=axs[i].transAxes, fontsize=10,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```

> **a)** Elabora un programa que genere polinomios P(x) de grado n = 0, 1, 2, 3, · · · , 12 que ajusten los datos anteriores. Asigna el nodo x = 0 al trimestre 1 del año 2018, el nodo x = 1 el trimestre 2 de 2018 y así sucesivamente. Para cada polinomio de ajuste calcula el promedio de los residuales.

```python
# Leer el csv
df = pd.read_csv("crecimiento-PBI.csv")

# Renombrar el df
PIB = df.rename(columns={'Porcentaje de crecimiento  del Producto Interno Bruto':'AnioTrimestre', 'Unnamed: 1':'CrecimientoPIB'})

#Extraer el valor del PIB
ys = ( PIB['CrecimientoPIB']).tolist()

#Crear df de interés
data = {'Xs': range(1,len(PIB)+1),
        'Ys': range(1,len(PIB)+1)}
nodos = pd.DataFrame(data)

for i in range(len(PIB)):
    nodos.loc[i, 'Ys'] = ys[i]


#Función para generar polinomios
def generar_polinomios_grado_n(x, y, n):
    """
    Genera polinomios de grado 0 hasta N que se ajustan a un conjunto específico de puntos.

    Parámetros:
    - x: Lista de valores x de los puntos.
    - y: Lista de valores y de los puntos.
    - n: Grado máximo del polinomio.

    Devuelve:
    - Lista de coeficientes de los polinomios generados.
    """
    coeficientes = []
    for grado in range(n + 1):
        # Ajuste polinómico de mínimos cuadrados
        coef = np.polyfit(x, y, grado)
        coeficientes.append(coef)

    return coeficientes

#Función para evaluar los polinomios
def evaluar_polinomio(coeficientes, x):
    """
    Evalúa un polinomio dados sus coeficientes y un valor x.

    Parámetros:
    - coeficientes: Lista de coeficientes del polinomio.
    - Xs: Valor a evaluar.

    Devuelve:
    - Valor y correspondiente al polinomio evaluado en x.
    """
    return np.polyval(coeficientes, x)

# Puntos del ejercicio
puntos_x = nodos['Xs'].tolist()
puntos_y = nodos['Ys'].tolist()

# Grado máximo del polinomio
grado_maximo = 12

# Generar polinomios
coeficientes_polinomios = generar_polinomios_grado_n(puntos_x, puntos_y, grado_maximo)

# Mostrar resultados
for i, coef in enumerate(coeficientes_polinomios):
    print(f"Coeficientes del polinomio de grado {i}: {coef}")

# Grados de los polinomios a ajustar
grados = range(0,12,1)

# Inicializar una lista para almacenar los promedios de residuales
promedios_residuales = []

# Iterar sobre los diferentes grados
for grado in grados:
    # Ajustar el polinomio de grado n
    coeficientes = np.polyfit(puntos_x, puntos_y, grado)

    # Calcular los valores ajustados (polinomio evaluado en xi)
    fy = np.polyval(coeficientes, puntos_x)

    # Calcular los residuales
    residuales = puntos_y - fy

    # Calcular el promedio de los residuales
    promedio_residual = np.mean(np.abs(residuales))

    # Almacenar el promedio de residuales en la lista
    promedios_residuales.append(promedio_residual)

    # Imprimir los promedios de residuales
for i, grado in enumerate(grados):
    print(f'Grado {grado}: Promedio de Residuales = {promedios_residuales[i]}')


print('El mínimo de residuales es: ',min(promedios_residuales))
```

El archivo “industria-construccion.cvs” contiene los datos del valor de la producci ́on (en miles de pesos) de la industria de la construccion del paıs, de enero de 2006 a marzo de 2022. Para obtener el valor total acumulado de la producci ́on debida a la construccion de manera aproximadamente continua podemos calcular una integral numerica. Entonces, realiza lo siguiente:

##a) Elabora un programa que calcule la integral num ́erica por medio de una regla de integracion compuesta. Utiliza la regla de integraci ́on compuesta de tu preferencia; asigna el nodo x = 0 al mes de enero de 2006, el nodo x = 1 al mes de febrero de 2006, y ası sucesivamente.

```python
import pandas as pd
import numpy as np

# Cargar datos
pobmex = pd.read_csv('industria-construccion.csv', thousands=',')
pobmex.head()

# Obtener los valores de la columna de producción
produccion = pobmex['Indicadores_economicos'].values

# Definir la función a integrar
def f(x):
    # Asumiendo que la frecuencia de los datos es mensual
    index = int(x)
    return produccion[:index + 1].sum()

# Calcular la integral numérica utilizando la regla del trapecio compuesta
def integral_trapecio_compuesta(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        result += f(a + i * h)
    result *= h
    return result

# Definir los límites de integración para el año 2006 (primeros 12 meses)
a = 0  # primer mes de 2006
b = 11  # último mes de 2006

# Número de subintervalos para la regla compuesta
n = b - a

# Calcular la integral numérica
valor_acumulado = integral_trapecio_compuesta(f, a, b, n)

# Imprimir resultados
print(f"Intervalo de integración: Desde el mes {a + 1} (enero de 2006) hasta el mes {b + 1} (diciembre de 2006)")
print(f"Número de subintervalos utilizados: {n}")
print(f"El valor total acumulado de la producción debida a la construcción en 2006 es aproximadamente: {valor_acumulado:.2f} miles de pesos.")
```
