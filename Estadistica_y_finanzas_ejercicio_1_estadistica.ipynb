```python
import numpy as np
import matplotlib.pyplot as plt

# Definir valores de t
t = np.linspace(0, 5, 500)

# Parámetros dados
lambda_ = 1
gamma = 0.5

# Función de supervivencia S(t)
S_t = np.exp(-lambda_ * t - (gamma * t**2) / 2)

# Función de riesgo h(t)
h_t = lambda_ + gamma * t

# Graficar S(t)
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.plot(t, S_t, label="S(t)")
plt.title("Función de Supervivencia S(t)")
plt.xlabel("t")
plt.ylabel("S(t)")
plt.grid(True)

# Graficar h(t)
plt.subplot(1, 2, 2)
plt.plot(t, h_t, label="h(t)", color='r')
plt.title("Función de Riesgo h(t)")
plt.xlabel("t")
plt.ylabel("h(t)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

1. $\textbf{Números de Catalan:}$

   Los números de Catalan se $C_{n}$ son una secuencia de numeros enteros $1, 1, 2, 5, 14, 42, 132$ que juegan un papel importante en la mecanica cúantica  y la teoría de los sistemas desordenados. (Eran fundamentales para la prueba de Eugene Wigner de la llamada ley del semicírculo). Se definen por:

   \begin{gather*}
   C_{0}=1 , ~~~C_{n+1}=\frac{4n+2}{n+2}C_{n}
   \end{gather*}

   Escribe un programa que imprima todos los numeros de Catalan $C_{n}$ menores o        iguales a $1 \times 10^{9}$ (mil millones).



```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

def binomial_coefficient(n,k):
    return factorial(n) // (factorial(k)*factorial(n-k))

def catalan_number(n):
    return binomial_coefficient(2 * n, n) // (n + 1)

limit = 10**9
n = 0

while catalan_number(n) <= limit:
    print(catalan_number(n))
    n += 1
```

$2.\textbf{La constante de Madelung}$:

En física de la materia condensada, la $\textit{la constante de Madelung}$ da el potencial eléctrico total que tiene un átomo en un sólido; y depende de las cargas de los otros átomos cercanos y de sus ubicaciones.

Por ejemplo el cristal de cloruro de sodio sólido (sal de mesa), tine átomos dispuestos en una red cúbica, con  átomos de sodio y cloro alternados, teniendo los de sodio una carga positiva $+e$ y los del cloro una negativa $-e$ (donde $e$ es la carga del electrón). Si etiquetamos cada posicion en la red con tres coordenadas enteras $(i,j,k)$ entonces los átomos de sodio caen en posiciones donde $i+j+k$ es par, y los atomos de cloro en posiciones donde $i + j + k$ es impar.

Consideremos un atomo de sodio en el origen, $i.e.  i = j = k = 0$, y calculemos la
$\textit{constante de Madelung}$. Si el espaciado de los atomos en la red es $a$, entonces la distancia desde el origen al atomo en la posición $(i, j, k)$ es:

$$\begin{gather}
\sqrt{(ia)^{2}+(ja)^{2}+(ka)^{2}} = a\sqrt{i^{2}+j^{2}+k^{2}},
\end{gather}$$

y el potencial en el origen creado por tal átomo es:

$$\begin{gather}
V(i,j,k)=\pm \frac{e}{4\pi\epsilon_{o}a\sqrt{i^{2}+j^{2}+k^{2}}},
\end{gather}$$

Siendo $\epsilon_{o}$ la permitividad del vacío y el signo de la expresión se toma dependiendo de si $i+j+k$ es par o impar. Así entonces, el potencial total que siente el átomo de sodio es la suma de esta cantidad sobre todos los demas átomos. supongamos  una caja cúbica alrededor del átomo de sodio en el origen, con $L$ átomos en todas las  direcciones, entonces:

$$\begin{gather}
V_{total}= \sum^{L}_{i,j,k=-L} V(i,j,k) = \frac{e}{4\pi\epsilon_{o}a}M,~~~~con ~~(i,j,k \neq 0)
\end{gather}$$

Donde $M$ es la constante de Madelung (al menos aproximadamente).

Tecnicamente, la constante de Madelung es el valor de $M$ cuando $L → \infty$, pero se puede obtener una buena aproximacion simplemente usando un valor grande de  $L$.





```python
import numpy as np

def madelung_constant(L):
    epsilon_0 = 8.854187817e-12 #Permitividad del vacio (F/m)
    e = -1.602176634e-19 #Carga del electron (C)
    a = 5.63e-10 #Espaciado de los átmo en el cloruro de sodio (m)

    total_potential = 0

    for i in range(-L, L + 1):
        for j in range(-L, L + 1):
            for k in range(-L, L + 1):
                if i != 0 or j != 0 or k != 0:
                    r = np.sqrt(i**2 + j**2 + k**2)
                    if (i +j +k) % 2 == 0:
                        total_potential += e / (4 * np.pi * epsilon_0 * a * r)
                    else:
                        total_potential -= e / (4* np.pi * epsilon_0 * a * r)

    return total_potential

L = 100 #Puedes variar el parametro
madelung = madelung_constant(L)
print("La constante de Madelung para el cloruro de sodio aproximadamente es es:"
f'{madelung:.2f}')
#Con valores de L>100 el programa tarda mas de 5min .
#El valor de la constante es mas cercano al real conforme L se mas grande
```

$(c)$ La probabilidad de que para una moneda no sesgada, lanzada $n$ veces, salga
aguila $k$ veces es:
$$\begin{gather}
p(k|n) = \frac{\begin{pmatrix}
n \\
k
\end{pmatrix}}{2^{n}}
\end{gather}$$

Escribe un programa para calcular:

$i.$ la probabilidad total de que una moneda lanzada 100 veces, salga águila exactamente 60 veces y

$ii.$ la probabilidad de que salga aguila 60 veces o más.

    

```python
#a)
def coeficiente_binomial(n,k):
    if k == 0:
        return 1
    if k > n:
        return 0


    result = 1
    for i in range(1, k + 1):
        result *= (n - i +1)
        result //= i

    return result

n = 5
k = 3

resultado = coeficiente_binomial(n, k)
print(f"coeficiente binomial c({n}, {k})= {resultado}")
```

$4) \textbf{Números primos}$

Es posible escribir un programa bastante rápido para calcular números primos siguiendo las siguientes observaciones:

$(a)$ Un número $n$ es primo, si no tiene factores primos menores que $n$. Por lo tanto, solo necesitamos comprobar si es divisible por otros números primos.  

$(b)$ Si un numero $n$ no es primo y tiene un factor $r$, entonces $n = rs$, donde $s$ también es un factor. Si $r ≥\sqrt{n}$ entonces $n = rs ≥\sqrt{ns}$, lo que implica que $s ≤\sqrt{n}$. Es decir, cualquier $\textbf{no primo}$ debe tener factores, y por lo tanto también factores primos, menores o iguales a $\sqrt{n}$. Por lo tanto, para determinar si un número es primo, tenemos que verificar sus factores primos solo hasta $\sqrt{n}$; incluso si no hay ninguno, entonces el número es primo.

$(c)$ Si encontramos un $\textbf{factor primo menor que}$ $\sqrt{n}$$\textbf{entonces sabemos que el número no es primo}$ y, por lo tanto, no hay necesidad de realizar mas comprobaciones; podemos abandonar este numero y pasar al siguiente.

$\textbf{Escribe un programa que encuentre todos los números primos hasta 10 000 (diez mil)}$. Crea una lista para almacenar los numeros primos, que comience con solo un número primo (el 2). Luego, para cada número $n$ del $3$ al $10 000$, verifica si el número es divisible por alguno de los numeros primos de la lista hasta $\sqrt{n}$. En cuanto encuentres un factor primo, puedes dejar de comprobar el resto de numeros (pues ya sabrías que $n$ no es primo). Si no encuentras factores primos iguales o menores a $\sqrt{n}$, entonces $n$ es primo y habría que agregarlo a la lista. Puedes imprimir la lista completa al final del programa, o bien, ir imprimiendo los números individualmente a medida que los vayas encontrando.


```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
        return True

for n in range(1, 10000):
    if es_primo(n):
        print(n)

```

```python
import math

# Inicializamos una lista con el primer número primo (2).
primos = [2]

# Iteramos desde 3 hasta 10,000 con saltos de 2 (solo números impares).
for n in range(3, 10001, 2):
    es_primo = True

    # Verificamos si n es divisible por algún número primo menor o igual a su raíz cuadrada.
    for primo in primos:
        if primo > math.isqrt(n):
            break  # No es necesario seguir verificando números primos más grandes.
        if n % primo == 0:
            es_primo = False
            break

    # Si n es primo, lo agregamos a la lista.
    if es_primo:
        primos.append(n)

# Imprimimos la lista de números primos encontrados.
for primo in primos:
    print(primo)
```
